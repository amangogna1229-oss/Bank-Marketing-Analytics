```python
bank_df = spark.read.table("silver_bank_marketing")

channel_df = spark.read.table("silver_channel")

customer_df = spark.read.table("silver_customer")

region_df = spark.read.table("silver_region")
```

## GOLD TABLE 1: Customer Overview

Total Customers \
Average Monthly Income \
Average Credit Score \
Average Account Tenure \
Total Transaction Amount 

```python
from pyspark.sql.functions import *

gold_customer_overview = bank_df.groupBy("region","customer_segment","age_group","income_band","risk_level","housing","loan").agg(
    countDistinct("customer_id").alias("total_customers"),
    round(avg("monthly_income"),2).alias("avg_income"),
    round(avg("credit_score"),2).alias("avg_credit_score"),
    round(avg("account_tenure_years"),2).alias("avg_account_tenure"),
    round(sum("last_transaction_amount"),2).alias("total_transaction_amount")
)
display(gold_customer_overview)
```

```python
gold_customer_overview.write \
.mode("overwrite") \
.format("delta") \
.option("mergeSchema", "true") \
.saveAsTable("gold_customer_overview")
```

## GOLD TABLE 2: Campaign Performance

Total Contacts \
Successful Campaigns \
Success Rate \
Average Duration \
Average Engagement Score

```python
from pyspark.sql.functions import *

gold_campaign_performance = bank_df.groupBy("region","channel","contact","job","education").agg(
    count("*").alias("total_contacts"),
    sum(when(col("y") == "yes", 1).otherwise(0)).alias("successful_campaigns"),
    round(avg("duration"),2).alias("avg_duration"),
    round(avg("engagement_score"),2).alias("avg_engagement_score")
)
display(gold_campaign_performance)
```

```python
gold_campaign_performance.write \
.mode("overwrite") \
.format("delta") \
.option("mergeSchema", "true") \
.saveAsTable("gold_campaign_performance")
```

##   GOLD TABLE 3: Risk & Revenue Insight


```python
from pyspark.sql.functions import *
gold_risk_revenue = bank_df.groupBy("risk_level","customer_segment","region","income_band").agg(
    count("customer_id").alias("total_customers"),
    sum("monthly_income").alias("total_monthly_income"),
    round(avg("monthly_income"),2).alias("avg_monthly_income"),
    round(avg("credit_score"),2).alias("avg_credit_score"),
    round(sum("last_transaction_amount"),2).alias("total_transaction_amount")
)
display(gold_risk_revenue)
```

```python
gold_risk_revenue.write \
.mode("overwrite") \
.format("delta") \
.option("mergeSchema", "true") \
.saveAsTable("gold_risk_revenue")
```
